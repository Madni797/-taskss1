```python
import pandas as pd

# Load the uploaded file
iris = pd.read_csv("iris-id.csv")

# Confirm it's loaded
print(iris.head())
```

```python
print("shape:", iris.shape)
print("column Names:",iris.columns.tolist())

print("\nINFO:")
print(iris.info())

print("\nDESCRIPTION:")
print(iris.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(iris, hue='target')
plt.suptitle('scatter plot of Iris features',y=1.02)
plt.show()
```

```python
iris.hist(edgecolor='black',figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris.drop('species', axis=1), orient='h')
plt.title("Boxplots of Iris Features")
plt.show()
```
